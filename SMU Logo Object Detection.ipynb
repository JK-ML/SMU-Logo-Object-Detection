```python
# Define paths
label_dir = "/content/drive/MyDrive/CS604 Computer Vision/CV Project/Dataset/Object Detection Data"
train_dir = "/content/drive/MyDrive/CS604 Computer Vision/CV Project/Dataset/Object Detection Data/data/labels/train"
val_dir = "/content/drive/MyDrive/CS604 Computer Vision/CV Project/Dataset/Object Detection Data/data/labels/val"
test_dir = "/content/drive/MyDrive/CS604 Computer Vision/CV Project/Dataset/Object Detection Data/data/labels/test"

# Initialize an empty list to store filenames
annotated_files = []
non_annotated_files = []

# Iterate through files in the directory
for filename in os.listdir(label_dir):
    file_path = os.path.join(label_dir, filename)
    # Check if it's a file and not a directory
    if os.path.isfile(file_path):
        # Read the content of the file
        with open(file_path, 'r') as file:
            content = file.read()
            # Check if content length is greater than 0
            if len(content.strip()) > 0:
                # Append filename to the list
                annotated_files.append(filename)
            else:
                non_annotated_files.append(filename)

num_train_non_annotated = int(0.6 * len(non_annotated_files))
num_val_non_annotated = int(0.2 * len(non_annotated_files))
num_test_non_annotated = len(non_annotated_files) - num_train_non_annotated - num_val_non_annotated

num_train_annotated = int(0.6 * len(annotated_files))
num_val_annotated = int(0.2 * len(annotated_files))
num_test_annotated = len(annotated_files) - num_train_annotated - num_val_annotated


random.shuffle(annotated_files)
random.shuffle(non_annotated_files)
```

```python
all_annotated = annotated_files + non_annotated_files
```

# Move annotated & non annotated label text files into respective train,val,test with 60%,20%,20% ratio

```python
# Copy images and annotations to train, val, and test directories
for i, label_file in enumerate(annotated_files):
    if i < num_train_annotated:
        copyfile(os.path.join(label_dir, label_file), os.path.join(train_dir, label_file))
    elif i < num_train_annotated + num_val_annotated:
        copyfile(os.path.join(label_dir, label_file), os.path.join(val_dir, label_file))
    else:
        copyfile(os.path.join(label_dir, label_file), os.path.join(test_dir, label_file))

# Copy images and annotations to train, val, and test directories
for i, label_file in enumerate(non_annotated_files):
    if i < num_train_non_annotated:
        copyfile(os.path.join(label_dir, label_file), os.path.join(train_dir, label_file))
    elif i < num_train_non_annotated + num_val_non_annotated:
        copyfile(os.path.join(label_dir, label_file), os.path.join(val_dir, label_file))
    else:
        copyfile(os.path.join(label_dir, label_file), os.path.join(test_dir, label_file))
```
